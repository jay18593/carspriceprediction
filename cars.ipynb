```python
avg_hp=usedcars["horsepower"].astype("float").mean()
```

```python
avg_hp
```

```python
usedcars["horsepower"].replace(np.nan,avg_hp,inplace=True)
```

```python
avg_peakrpm=usedcars["peak-rpm"].astype("float").mean()
```

```python
usedcars["peak-rpm"].replace(np.nan,avg_peakrpm,inplace=True)
```

```python
usedcars["num-of-doors"].value_counts()
#here we use value_counts() method to see how many types of different value counts for each variable we have,
#we will replace null values with the most common variable which occured mpore times in a cell.
```

```python
usedcars["num-of-doors"].value_counts().idxmax()
#if we use idxmax() method in the end then it will give you the most occured value of that column.
```

```python
usedcars["num-of-doors"].replace(np.nan,"four",inplace=True)
```

```python
usedcars.dropna(subset=["price"], axis=0, inplace=True)
#axis=0 is for rows.
#here is the end of dealing with missing values, Now we do not have any missing data in our dataset.
```

```python
# reset index, because we droped two rows
usedcars.reset_index(drop=True, inplace=True)
```

```python
usedcars.dtypes
```

```python
#convert data types to proper format
usedcars[["bore","stroke"]]=usedcars[["bore","stroke"]].astype("float")
```

```python
usedcars[["normalized-losses"]]=usedcars[["normalized-losses"]].astype("int")
usedcars[["price"]]=usedcars[["price"]].astype("float")
usedcars[["peak-rpm"]]=usedcars[["peak-rpm"]].astype("float")
```

```python
usedcars.rename(columns={'highway-mpg':'highway-l/100'},inplace=True)
```

```python
usedcars.columns
```

```python
usedcars["highway-l/100"]=235/usedcars["highway-l/100"]
```

```python
usedcars.head()
```

```python
usedcars["city-l/100"]=235/usedcars["city-mpg"]
```

```python
usedcars.head()
```

```python
#data normalization: It is a process to convert values of variables in a range of 0 to 1 so while we do analysis,
#or make the model, results doesn"t get skwedor get affected because of their big value range.
usedcars["length"]=usedcars["length"]/usedcars["length"].max()
```

```python
usedcars["width"]=usedcars["width"]/usedcars["width"].max()
```

```python
usedcars["height"]=usedcars["height"]/usedcars["height"].max()
```

```python
usedcars["horsepower"]=usedcars["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(usedcars["horsepower"])

#set x/y lables and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins=np.linspace(min(usedcars["horsepower"]),max(usedcars["horsepower"]),4)
bins
```

```python
group_names=['low','medium','high']
```

```python
#now we need cut function...
usedcars["horsepower-binned"]=pd.cut(usedcars["horsepower"], bins, labels=group_names, include_lowest=True)
usedcars[["horsepower","horsepower-binned"]].head(20)
```

```python
usedcars["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, usedcars["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#now we need dummy variables for fuel-type column bcz regression analysis can be done with numeric values only so we need to
#convert fuel-type column in to dummy variable.
dummy_variable_1=pd.get_dummies(usedcars["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas','fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1"
usedcars = pd.concat([usedcars, dummy_variable_1], axis=1)
# drop original column "fuel-type" from "df"
usedcars.drop("fuel-type", axis=1, inplace=True)
```

```python
dummy_variable_2=pd.get_dummies(usedcars["aspiration"])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns={'aspiration':'std','aspiration':'turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
usedcars = pd.concat([usedcars,dummy_variable_2], axis=1)
```

```python
usedcars.drop("aspiration", axis=1, inplace=True)
```

```python
usedcars.to_csv('clean_usedcars.csv')
```
